```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1,1), (1,2), (1,3), (2,3), (3,4), (4,5), (1,5)])
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
print("Nodes: ", G.nodes())
print("Edge: ", G.edges())
print("Number of Nodes: ", G.number_of_nodes())
print("Numberof Edges: ", G.number_of_edges())
```

```python
print("Is the graph is connected? ", nx.is_connected(G))
```

```python
#Find the shortest path between 1 and 4
print("Print shortest Path between 1 and 4", nx.shortest_path(G, source = 1, target = 4))
```

```python
#Degree of each node
degree_sequence = G.degree()
print("Degree of each node: ", degree_sequence)
```

```python
#Degree centrality
print("Degree of centrality: ", nx.degree_centrality(G))

#Clustering Coeficient
print("Clustering Coefficient:", nx.clustering(G))
```

```python
#Draw a subgraph by extracting nodes and edge
subgraph_nodes = [1,2,3]
subgraph = G.subgraph(subgraph_nodes)

nx.draw(subgraph, with_labels = True, font_weight = 'bold')
plt.show()
```

```python
#Duplicate the graph
G_copy = G.copy()


#Remove nodes
G_copy.remove_node(4)

nx.draw(G_copy, with_labels = True, font_weight = 'bold')
plt.show()
```

```python
#Remove nodes
G_copy.remove_edge(1, 5)

nx.draw(G_copy, with_labels = True, font_weight = 'bold')
plt.show()
```

```python
#Find the connected components
components = list(nx.connected_components(G_copy))
print("Print the connected component", components)

for i, components in enumerate(components):
  subgraph = G_copy.subgraph(components)
  plt.figure()

  nx.draw(subgraph, with_labels = True, font_weight = 'bold')
  plt.title(f"Connected components {i+1}")
  plt.show()
```
